```
# necessary modules
import pandas as pd
from sklearn.linear_model import LinearRegression
import math
import numpy as np

#dataset
dataset = pd.read_csv('/content/drive/My Drive/nasa/tourismhanif.csv')
data = dataset.rename(columns = {"Year":"year", "Accommodation services":"accomodation", "Food and beverage serving services":"food","Passenger transport services":"transportion", "Travel agencies and other reservation services": "agency", "Cultural, sports and recreational services": "culacti","Country-specific tourism characteristic goods":"ctrgoods","Country-specific tourism characteristic services":"ctrservice","Total":"total"})


def create_predicted_tourism(data):
  """
  input: previous tourism dataset
  output: predicted tourism dataset
  
  """
  # years to be predicted
  pred_year = [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]
  
  # lists for predicted data to be stored
  accomodation = []
  food = []
  transportation = []
  agency = []
  cul_acti = []
  ctr_goods = []
  ctr_service = []
  total = []

  features = ['year']
  x = data[features]
  
  # model for predicting accomodation expenditure
  acc_model = LinearRegression()
  acc_model.fit(x,data.accomodation)
  
  # model for predicting Food and beverage serving services expenditure
  food_model = LinearRegression()
  food_model.fit(x,data.food)
  
  # model for predicting Passenger transport services expenditure
  tran_model = LinearRegression()
  tran_model.fit(x,data.transportion)
  
  # model for predicting Travel agencies and other reservation services expenditure
  agency_model = LinearRegression()
  agency_model.fit(x,data.agency)
  
  # model for predicting Cultural, sports and recreational services expenditure
  cul_model = LinearRegression()
  cul_model.fit(x,data.culacti)
  
  # model for predicting Country-specific tourism characteristic goods expenditure
  goods_model = LinearRegression()
  goods_model.fit(x,data.ctrgoods)
  
  # model for predicting Country-specific tourism characteristic services expenditure
  service_model = LinearRegression()
  service_model.fit(x,data.ctrservice)
  
  # model for predicting total expenditure
  total_model = LinearRegression()
  total_model.fit(x,data.total)
  
  for year in pred_year:
    prediction = 0
    prediction = acc_model.predict([[year]])
    accomodation.append(math.ceil(prediction[0]))
    
    prediction = food_model.predict([[year]])
    food.append(math.ceil(prediction[0]))
    
    prediction = tran_model.predict([[year]])
    transportation.append(math.ceil(prediction[0]))
    
    prediction = agency_model.predict([[year]])
    agency.append(math.ceil(prediction[0]))
    
    prediction = cul_model.predict([[year]])
    cul_acti.append(math.ceil(prediction[0]))
    
    prediction = goods_model.predict([[year]])
    ctr_goods.append(math.ceil(prediction[0]))
    
    prediction = service_model.predict([[year]])
    ctr_service.append(math.ceil(prediction[0]))
    
    prediction = total_model.predict([[year]])
    total.append(math.ceil(prediction[0]))
    
    
  predicted_future = pd.DataFrame({"Year":pred_year, "Accommodation services": accomodation,
                                   "Food and beverage serving services":food, "Passenger transport services": transportation,
                                   "Travel agencies and other reservation services":agency, 
                                   "Cultural, sports and recreational services":cul_acti, 
                                   "Country-specific tourism characteristic goods": ctr_goods, 
                                   "Country-specific tourism characteristic services":ctr_service, "Total": total})
  
  predicted_future.to_csv("/content/drive/My Drive/nasa/future_tourism_expenditure.csv")
  
  
create_predicted_tourism(data)
```
